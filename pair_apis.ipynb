```python
import requests

# Pedimos al usuario el nombre o parte del nombre
nombre = input("Escribe el nombre (o parte del nombre) de un personaje: ").lower()

# Hacemos la búsqueda en el API
url = "https://swapi.py4e.com/api/people/?search=" + nombre
respuesta = requests.get(url)

# Convertimos la respuesta en formato JSON
datos_starwars = respuesta.json()

# Recorremos los personajes encontrados
for personaje in datos_starwars["results"]:
    print(personaje["name"], "-", personaje["gender"])
```

```python
import requests
from IPython.display import Image, display

api_key = "live_WPOnfWVi8KsZFVeZioTg6AV0ysfehoUNPGbWpx74IT6aKtSrfvr3xQZz9pG7O4ES"

# Hacemos la petición a TheCatAPI
url = "https://api.thecatapi.com/v1/images/search"
headers = {"x-api-key": api_key}
params = {"mime_types": "gif"}  # Para pedir GIFs

response = requests.get(url, headers=headers, params=params)
data = response.json()

# Obtenemos la URL de la imagen
cat_url = data[0]["url"]
print("Imagen del gato:", cat_url)

# Mostramos ("pintamos") la imagen
display(Image(url=cat_url))
```
