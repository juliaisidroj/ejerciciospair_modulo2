```python
import requests

# Pedimos al usuario el nombre o parte del nombre
nombre = input("Escribe el nombre (o parte del nombre) de un personaje: ").lower()

# Hacemos la búsqueda en el API
url = "https://swapi.py4e.com/api/people/?search=" + nombre
respuesta = requests.get(url)

# Convertimos la respuesta en formato JSON
datos_starwars = respuesta.json()

# Recorremos los personajes encontrados
for personaje in datos_starwars["results"]:
    print(personaje["name"], "-", personaje["gender"])
```
